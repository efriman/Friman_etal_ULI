```python
combined = pd.DataFrame()
nameslist = ["RAD21", "CTCF", "WAPL", "RING1B"]
dflist = [[RAD21_UT_short, RAD21_UT_long, RAD21_AID_short, RAD21_AID_long],
         [CTCF_UT_short, CTCF_UT_long, CTCF_AID_short, CTCF_AID_long],
         [WAPL_UT_short, WAPL_UT_long, WAPL_AID_short, WAPL_AID_long],
         [RING1B_WT_short, RING1B_WT_long, RING1B_KO_short, RING1B_KO_long]]
for i in range(len(nameslist)):
    dflist[i][0]["Range"] = "short"
    dflist[i][1]["Range"] = "long"
    dflist[i][2]["Range"] = "short"
    dflist[i][3]["Range"] = "long"
    dflist[i][0]["Condition"] = "WT"
    dflist[i][1]["Condition"] = "WT"
    dflist[i][2]["Condition"] = "Treatment"
    dflist[i][3]["Condition"] = "Treatment"
    concat = pd.concat(dflist[i])
    concat["Degron"] = nameslist[i]
    combined = pd.concat([combined, concat], axis=0)
    
combined["Factor"] = combined.ID_Factor.str.split('_', n=1).str[1]
```

```python
combined["F_mean"] = combined.groupby(["Range", "Condition", "Degron"])["F_value"].transform("mean")
```

```python
combined["F_normalised"] = combined["F_value"] / combined["F_mean"]
```

```python
#Manual annotation of groups
TF = TFs["Factor"].tolist() + ["SMAD2/3", "SMAD2-3", 'ISL1-2']
Cofactors = Cofactor["Factor"].tolist() + ["ZNF143", "CFP1", "RPB1", "KMT2D", "CHD2", "SUPT5H"]
Cohesin = ['CTCF', 'SMC1A', 'RAD21', 'STAG1', 'STAG2', 'SMC3', 'WAPL', 'NIPBL']
Polycomb = ['PHC1', 'RING1B', 'RING1A', 'RING1B', 'EED', 'EZH2', 'CBX2', 'SUZ12', 
'CBX1', 'CBX7', 'NONO', 'RBP1', 'MTF2', 'JARDI2', 'PCGF2', 'RYBP', 'RBBP5', 'WDR5',
'PCGF6', 'CBX6', 'PCGF1', 'CBX5', 'PHF19', 'CBX8', 'CBX3', 'JARID2', 'EPOP', 'BAP1']
Repressive = ['CHD4', 'TRIM28', 'SINHCAF', 'TET1', 'TET3', 'SIN3A', 'DNMT3B', 'REST', 'HEXIM1']
```

```python
combined["Group"] = "Other"
combined.loc[combined["Factor"].isin(TF),"Group"] = "TF"
combined.loc[combined["Factor"].isin(Cofactors),"Group"] = "Co-factors"
combined.loc[combined["Factor"].isin(Repressive),"Group"] = "Repressive"
combined.loc[combined["Factor"].isin(Polycomb),"Group"] = "Polycomb"
combined.loc[combined["Factor"].isin(Cohesin),"Group"] = "Cohesin"
```

```python
palette2 = {f:'whitesmoke' for f in set(combined["Group"])}
for f in combined.loc[combined["Group"] == "TF","Group"]:
    palette2[f] = '#4daf4a' 
for f in combined.loc[combined["Group"] == "Co-factors","Group"]:
    palette2[f] = '#e41a1c' 
for f in combined.loc[combined["Group"] == "Cohesin","Group"]:
    palette2[f] = 'orange'
for f in combined.loc[combined["Group"] == "Polycomb","Group"]:
    palette2[f] = '#377eb8'
for f in combined.loc[combined["Group"] == "Repressive","Group"]:
    palette2[f] = 'cyan'
```

```python
sns.set(style="ticks", font="Arial")
sns.set_context("notebook", rc={"font.size":1,"axes.titlesize":20, 
                                "axes.labelsize":12, "xtick.labelsize": 12,
                                "ytick.labelsize": 12}) 
plt.figure(figsize=(5.2,2.5))
data = pd.pivot(combined.loc[(combined["Range"]=="short") &
                             (combined["Degron"]=="RAD21"),
                             ["ID_Factor", "F_normalised", "Condition", "Group"]], 
                index=["ID_Factor", "Group"], 
                columns="Condition", 
                values=["F_normalised"]).droplevel(0, axis=1).reset_index()
data["Treatment/WT"] = data["Treatment"] / data["WT"]
g = sns.stripplot(data=data, y="Treatment/WT", x="Group", hue="Group", palette=palette2, legend=False,
                  edgecolor="black", linewidth=0.5, zorder=1, size=3,
                  order=["Cohesin", "Polycomb", "Co-factors", "TF", "Repressive", "Other"])
g.set(title='RAD21-AID short-range', ylabel="Normalised F-value [AID/UT]", xlabel = "")
plt.tight_layout()
plt.axhline(1,alpha=1, dashes=(1, 1), color="black")
plt.setp(g.artists, edgecolor = 'k')
plt.setp(g.lines, color='k')
sns.despine(trim=True, offset=0)
plt.tight_layout()
plt.savefig("plots/RAD21_shortrange_AID.svg")
```
